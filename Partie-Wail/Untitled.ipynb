```python
vehicules = pd.read_csv("vehicules-2023.csv", sep=";")
caract = pd.read_csv("caract-2023.csv", sep=";")
usagers = pd.read_csv("usagers-2023.csv", sep=";")

# Jointure entre caractéristiques et véhicules sur Num_Acc
caract_vehicules = pd.merge(caract, vehicules, on="Num_Acc", how="inner")

# Jointure avec usagers sur Num_Acc et id_vehicule
data_finale = pd.merge(caract_vehicules, usagers, on=["Num_Acc", "id_vehicule"], how="inner")

# Afficher un aperçu des données fusionnées
print(data_finale.head())

# Sauvegarder dans un nouveau fichier CSV si besoin
data_finale.to_csv("data_fusionnee.csv", index=False, sep=";") 
variables_explicatives = [
    "dep", "com", "agg", "catv", "motor",
    "sexe", "an_nais", "lum", "atm", "occutc",
    "place", "trajet", "secu1", "secu2", "secu3","grav"
]

# Filtrer les colonnes selon la liste
data_filtree = data_finale[variables_explicatives]

# Afficher un aperçu des données filtrées
print(data_filtree.head())
```

```python
from datetime import datetime
# Filtrer les colonnes selon la liste
data_filtree = data_finale[variables_explicatives]

# Calculer l'âge
annee_actuelle = datetime.now().year
data_filtree = data_filtree.copy()  # Créer une copie explicite pour éviter SettingWithCopyWarning
data_filtree['age'] = annee_actuelle - data_filtree['an_nais']

# Supprimer la colonne 'an_nais' car elle n'est plus nécessaire
data_filtree = data_filtree.drop(columns=['an_nais'])

# Afficher un aperçu des données
print(data_filtree.head())

# Sauvegarder les données
data_filtree.to_csv("data_explicatives_age.csv", index=False, sep=";")
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest

# 1. Appliquer IsolationForest pour détecter les outliers dans la variable 'age'
model = IsolationForest(contamination=0.01)  # contamination = proportion d'outliers attendus
data_filtree['outlier'] = model.fit_predict(data_filtree[['age']])

# 2. 'outlier' va valoir 1 pour les données normales et -1 pour les outliers
# Supprimer les outliers
data_no_outliers_iforest = data_filtree[data_filtree['outlier'] == 1]

# 3. Afficher le nombre d'outliers supprimés
outliers_count = data_filtree.shape[0] - data_no_outliers_iforest.shape[0]
print(f"Nombre d'outliers supprimés avec IsolationForest : {outliers_count}")

# 4. Visualiser les données avec boîte à moustaches avant suppression des outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x=data_filtree['age'])
plt.title("Boîte à Moustaches - Âge avant suppression des outliers")
plt.show()

# 5. Visualiser les données avec boîte à moustaches après suppression des outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x=data_no_outliers_iforest['age'])
plt.title("Boîte à Moustaches - Âge après suppression des outliers")
plt.show()

# 6. Distribution (Histogramme) avant suppression des outliers
plt.figure(figsize=(12, 6))
sns.histplot(data_filtree['age'], kde=True, bins=30)
plt.title("Distribution des Âges avant suppression des outliers")
plt.xlabel("Âge")
plt.ylabel("Fréquence")
plt.show()

# 7. Distribution (Histogramme) après suppression des outliers
plt.figure(figsize=(12, 6))
sns.histplot(data_no_outliers_iforest['age'], kde=True, bins=30)
plt.title("Distribution des Âges après suppression des outliers")
plt.xlabel("Âge")
plt.ylabel("Fréquence")
plt.show()

# 8. Comparaison des distributions avant et après suppression des outliers
plt.figure(figsize=(12, 6))

# Avant suppression des outliers
sns.histplot(data_filtree['age'], kde=True, color='blue', label='Avant suppression', bins=30)

# Après suppression des outliers
sns.histplot(data_no_outliers_iforest['age'], kde=True, color='red', label='Après suppression', bins=30)

plt.title("Comparaison des distributions des Âges avant et après suppression des outliers")
plt.xlabel("Âge")
plt.ylabel("Fréquence")
plt.legend()
plt.show()

data_filtree = data_filtree.drop(columns=['outlier'])
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Supposons que data_filtree soit votre DataFrame contenant les données

# Créer un tableau de contingence pour les variables 'dep' et 'com'
contingency_table = pd.crosstab(data_filtree['dep'], data_filtree['com'])

# Effectuer le test du chi-deux
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Affichage des résultats du test Chi-deux
print(f"Chi2: {chi2}")
print(f"P-value: {p}")

# Interprétation des résultats
if p < 0.05:
    print("Il existe une association significative entre 'dep' et 'com', suggérant une redondance.")
    print("Vous pouvez envisager de supprimer 'com'.")
    # Supprimer 'com' si redondant
    data_filtree = data_filtree.drop(columns=['com'])
    print("La variable 'com' a été supprimée en raison de sa redondance avec 'dep'.")
else:
    print("Il n'y a pas de relation significative entre 'dep' et 'com', 'com' contient des informations indépendantes de 'dep'.")
    print("La variable 'com' a été conservée.")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Étape 1: Vérifier et gérer les valeurs manquantes dans 'age'
# Remplacer les valeurs manquantes par la moyenne de la colonne 'age'
data_filtree['age'] = data_filtree['age'].fillna(data_filtree['age'].mean())

# Étape 2: Sélectionner la colonne 'age' pour appliquer le KMeans
age_data = data_filtree[['age']]

# Étape 3: Normalisation de la variable 'age' (c'est souvent une bonne pratique pour KMeans)
scaler = StandardScaler()
age_data_scaled = scaler.fit_transform(age_data)

# Méthode du coude pour déterminer le nombre optimal de clusters
inertia = []  # Liste pour stocker les valeurs d'inertie

# Tester différentes valeurs de k (par exemple de 1 à 10)
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(age_data_scaled)
    inertia.append(kmeans.inertia_)

# Tracer la courbe du coude
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Méthode du Coude')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# À partir du graphique du coude, choisissez le k optimal, qui correspond généralement au "genou" de la courbe.
# Par exemple, supposons que le nombre optimal de clusters soit k=3, vous pouvez l'appliquer ci-dessous.

# Étape 4: Appliquer KMeans avec 3 clusters (en supposant que k=3 soit le meilleur choix)
kmeans = KMeans(n_clusters=3, random_state=42)
data_filtree['age_class'] = kmeans.fit_predict(age_data_scaled) + 1  # Ajouter 1 pour avoir des classes 1, 2, 3

# Étape 5: Analyser la répartition des âges dans chaque classe
age_class_distribution = data_filtree.groupby('age_class')['age'].describe()

# Étape 6: Afficher les résultats
print(data_filtree[['age', 'age_class']].head())  # Afficher les premières lignes avec les classes de 'age'
print("\nDistribution des âges dans chaque classe :")
print(age_class_distribution)  # Afficher les statistiques descriptives par classe



```
